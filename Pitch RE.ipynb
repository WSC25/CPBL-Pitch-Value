```python
result = data.groupby(['球種', 'Outs', 'Balls', 'Strikes']).agg({'RunsScored': ['count', 'sum']})
result
```

```python
result.reset_index(inplace=True)
```

```python
result
```

```python
result['Expected_Runs']=result['RunsScored']['sum']/result['RunsScored']['count']
result
```

```python
#result.to_excel('各球種球數出局數下的得分期望值.xlsx', index=True)
```

```python
def get_expected_runs(row):
    auto_pitch_type = row["球種"]
    outs = row["Outs"]
    balls = row["Balls"]
    strikes = row["Strikes"]
    
    matching_row = result[(result["球種"] == auto_pitch_type) & 
                          (result["Outs"] == outs) & 
                          (result["Balls"] == balls) & 
                          (result["Strikes"] == strikes)]
    
    if not matching_row.empty:
        return matching_row.iloc[0]["Expected_Runs"]
    else:
        return None

data["Expected_Runs"] = data.apply(get_expected_runs, axis=1)

# 打印结果
data
```
