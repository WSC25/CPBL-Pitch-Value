```python
import pandas as pd
```

```python
excel_file_path3 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/球種回填(2018~2020).xlsx'
excel_file_path4 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2021_trackman.xlsx'
excel_file_path5 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2022_trackman.xlsx'
excel_file_path6 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2023_trackman1.xlsx'
excel_file_path7 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2023_trackman2.xlsx'
excel_file_path8 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/程式碼/Pitch Value/non_track_RE.xlsx'
sheet_name = 'Sheet1'
```

```python
track_18_20 = pd.read_excel(excel_file_path3, sheet_name=sheet_name)
track_2021 = pd.read_excel(excel_file_path4, sheet_name=sheet_name)
track_2022 = pd.read_excel(excel_file_path5, sheet_name=sheet_name)
track_2023_1 = pd.read_excel(excel_file_path6, sheet_name=sheet_name)
track_2023_2 = pd.read_excel(excel_file_path7, sheet_name=sheet_name)
non_track = pd.read_excel(excel_file_path8, sheet_name=sheet_name)
```

```python
col=['Date','球種','RunsScored','Outs','Balls','Strikes','Inning']
```

```python
track_18_20= track_18_20[col]
track_2021= track_2021[col]
track_2022= track_2022[col]
track_2023_1= track_2023_1[col]
track_2023_2= track_2023_2[col]
non_track=non_track[col]
```

```python
data3 = pd.concat([track_18_20, track_2021], ignore_index=True)
data4 = pd.concat([data3, track_2022], ignore_index=True)
data5 = pd.concat([data4, track_2023_1], ignore_index=True)
data6 = pd.concat([data5, track_2023_2], ignore_index=True)
data = pd.concat([data6, non_track], ignore_index=True)
```

```python
PitchType_items =data['球種'].unique()
PitchType_items
```

```python
#data = data.dropna()
data = data[data['球種'] != 'Undefined']
data = data[data['球種'] != 'Other']
```

```python
result = data.groupby(['球種', 'Outs', 'Balls', 'Strikes']).agg({'RunsScored': ['count', 'sum']})
result
```

```python
result.reset_index(inplace=True)
result
```

```python
result['Expected_Runs']=result['RunsScored']['sum']/result['RunsScored']['count']
result
```
