```python
import pandas as pd
```

```python
excel_file_path1 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Season_data/2024_non_trackman.xlsx'
excel_file_path2 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/程式碼/Pitch Value/各球種球數出局數下的得分期望值(投手).xlsx'
sheet_name = 'Sheet1'
```

```python
track_2024 = pd.read_excel(excel_file_path1, sheet_name=sheet_name)
RE = pd.read_excel(excel_file_path2, sheet_name=sheet_name)
```

```python
col=['pType','outs','balls','strikes','runsOnPlay','bName','pName','date','gameId','paRound','paOrder']
```

```python
track_2024=track_2024[col]
```

```python
name_mapping = {
    'SL':'Slider',
    'SI':'Sinker', 
    'FF':'Fastball', 
    'CH':'ChangeUp', 
    'CU':'Curveball', 
    'FO':'Splitter', 
    'FC':'Cutter',
    'KN':'Knuckle curve',
    'FS':'Splitter'
    }

# 新增一列"球員"，根据"Pitcher"列的值匹配中文名
track_2024 ['球種'] = track_2024 ['pType'].map(name_mapping)
```

```python
track_2024['PitchofPA'] = track_2024.groupby(['bName', 'pName', 'date', 'gameId', 'paRound', 'paOrder']).cumcount() + 1
```

```python
track_2024
```

```python
def get_expected_runs(row):
    auto_pitch_type = row["球種"]
    outs = row["outs"]
    balls = row["balls"]
    strikes = row["strikes"]
    
    matching_row = RE[(RE["球種"] == auto_pitch_type) & 
                          (RE["outs"] == outs) & 
                          (RE["balls"] == balls) & 
                          (RE["strikes"] == strikes)]
    
    if not matching_row.empty:
        return matching_row.iloc[0]["Expected_Runs"]
    else:
        return None

track_2024["Expected_Runs"] = track_2024.apply(get_expected_runs, axis=1)


track_2024
```

```python
track_2024['Expected_Runs'].fillna(0, inplace=True)
```

```python
track_2024.sort_values(by=['date','pName', 'bName', 'PitchofPA'], inplace=True)

# 计算每个球之间预期分数的变化
track_2024['RV'] =track_2024['Expected_Runs'].diff()

# 将PitchofPA为1的RV定为0
track_2024.loc[track_2024['PitchofPA'] == 1, 'RV'] = 0

# 显示处理后的数据
track_2024
```

```python
date_items =track_2024['date'].unique()
date_items
```

```python
rv_sum = track_2024.groupby(['pName', '球種'])['RV'].sum()

rv_sum
```

```python
rv_sum =track_2024.groupby(['pName', '球種'])['RV'].sum()

pitch_count = track_2024.groupby(['pName', '球種']).size()

result_df = pd.DataFrame({'PV_sum': rv_sum, 'PitchCount': pitch_count})

result_df
```

```python
result_df['PV/100']=(rv_sum/pitch_count)*100
```

```python
result_df.reset_index(inplace=True)
```
