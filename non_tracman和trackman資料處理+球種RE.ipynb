## 紀錄系統

```python
data_2024 = pd.read_excel(excel_file_path1, sheet_name=sheet_name)
data_2023 = pd.read_excel(excel_file_path2, sheet_name=sheet_name)
```

```python
non_track = pd.concat([data_2023, data_2024], ignore_index=True)
```

```python
non_track
```

```python
PitchType_items =non_track['pType'].unique()
PitchType_items
```

```python
non_track.pType.value_counts(dropna=False)
```

```python
name_mapping = {
    'SL':'Slider',
    'SI':'Sinker', 
    'FF':'Fastball', 
    'CH':'ChangeUp', 
    'CU':'Curveball', 
    'FO':'Splitter', 
    'FC':'Cutter',
    'KN':'Knuckle curve',
    'FS':'Splitter'
    }

# 新增一列"球員"，根据"Pitcher"列的值匹配中文名
non_track['球種'] = non_track['pType'].map(name_mapping)
```

```python
non_track.info()
```

```python
non_track_result = non_track.groupby(['球種', 'outs', 'balls', 'strikes']).agg({'runsOnPlay': ['count', 'sum']})
non_track_result
```

```python
non_track_result.reset_index(inplace=True)
non_track_result
```

## Trackman

```python
col=['Date','球種','PlayResult','RunsScored','Outs','Balls','Strikes','Inning']
```

```python
track_18_20 = pd.read_excel(excel_file_path3, sheet_name=sheet_name)
track_2021 = pd.read_excel(excel_file_path4, sheet_name=sheet_name)
track_2022 = pd.read_excel(excel_file_path5, sheet_name=sheet_name)
```

```python
track_18_20= track_18_20[col]
track_2021= track_2021[col]
track_2022= track_2022[col]
```

```python
data3 = pd.concat([track_18_20, track_2021], ignore_index=True)
data = pd.concat([data3, track_2022], ignore_index=True)
data
```

```python
PitchType_items =data['球種'].unique()
PitchType_items
```

```python
data = data.dropna()
data = data[data['球種'] != 'Undefined']
data = data[data['球種'] != 'Other']
```

```python
data.球種.value_counts(dropna=False)
```

```python
Trackman_result = data.groupby(['球種', 'Outs', 'Balls', 'Strikes']).agg({'RunsScored': ['count', 'sum']})
Trackman_result
```

```python
Trackman_result.reset_index(inplace=True)
Trackman_result
```

```python
#統一和紀錄系統欄位同名
new_column_names = {
    'Outs': 'outs',
    'Balls': 'balls',
    'Strikes': 'strikes',
    'RunsScored': 'runsOnPlay'
}
Trackman_result.rename(columns=new_column_names, inplace=True)
```

```python
Trackman_result
```

## 兩結果合併

```python
merged_df = pd.merge(non_track_result, Trackman_result, on=['球種', 'outs', 'balls', 'strikes'], suffixes=('_df1', '_df2'))

# 計算 'count' 和 'sum' 的總和
merged_df['total_count'] = merged_df['runsOnPlay_df1']['count'] + merged_df['runsOnPlay_df2']['count']
merged_df['total_sum'] =  merged_df['runsOnPlay_df1']['sum'] + merged_df['runsOnPlay_df2']['sum']
```

```python
merged_df.info()
```

```python
merged_df
```

```python
final_df = merged_df[['球種', 'outs', 'balls', 'strikes', 'total_count', 'total_sum']]
final_df
```

```python
final_df['Expected_Runs']=final_df['total_sum']/final_df['total_count']
final_df
```

```python
final_df.to_excel('各球種球數出局數下的得分期望值(投手).xlsx', index=True)
```
