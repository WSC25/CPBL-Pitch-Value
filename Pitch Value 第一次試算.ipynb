```python
import pandas as pd
```

```python
excel_file_path = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2024_trackman.xlsx'
excel_file_path2 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2023_trackman1.xlsx'
excel_file_path3 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2023_trackman2.xlsx'
excel_file_path1 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/各球種球數出局數下的得分期望值.xlsx'
sheet_name = 'Sheet1'
```

```python
track_2024 = pd.read_excel(excel_file_path, sheet_name=sheet_name)
track_2023_1 = pd.read_excel(excel_file_path2, sheet_name=sheet_name)
track_2023_2 = pd.read_excel(excel_file_path3, sheet_name=sheet_name)
RE = pd.read_excel(excel_file_path1, sheet_name=sheet_name)
```

```python
col=['Date','Pitcher','PitcherTeam','Batter','PitchofPA','TaggedPitchType','AutoPitchType','PlayResult','RunsScored','Outs','Balls','Strikes','Inning','Top/Bottom','Notes']
```

```python
RE
```

```python
track_2024= track_2024[col]
track_2023_1= track_2023_1[col]
track_2023_2= track_2023_2[col]
```

```python
data4 = pd.concat([track_2023_1, track_2023_2], ignore_index=True)
data = pd.concat([data4, track_2024], ignore_index=True)
```

```python
PitchType_items =RE['球種'].unique()
PitchType_items
```

```python
PitchType_items =data['TaggedPitchType'].unique()
PitchType_items
```

```python
def get_expected_runs(row):
    auto_pitch_type = row["TaggedPitchType"]
    outs = row["Outs"]
    balls = row["Balls"]
    strikes = row["Strikes"]
    
    matching_row = RE[(RE["球種"] == auto_pitch_type) & 
                          (RE["Outs"] == outs) & 
                          (RE["Balls"] == balls) & 
                          (RE["Strikes"] == strikes)]
    
    if not matching_row.empty:
        return matching_row.iloc[0]["Expected_Runs"]
    else:
        return None

data["Expected_Runs"] = data.apply(get_expected_runs, axis=1)

# 打印结果
data
```

```python
data['Expected_Runs'].fillna(0, inplace=True)
```

```python
data.info()
```

```python
data.sort_values(by=['Date','Pitcher', 'Batter', 'PitchofPA'], inplace=True)

# 计算每个球之间预期分数的变化
data['RV'] =data['Expected_Runs'].diff()

# 将PitchofPA为1的RV定为0
data.loc[data['PitchofPA'] == 1, 'RV'] = 0

# 显示处理后的数据
data
```

```python
rv_sum = data.groupby(['Pitcher', 'TaggedPitchType','PitcherTeam'])['RV'].sum()

rv_sum
```

```python
rv_sum =data.groupby(['Pitcher', 'PitcherTeam', 'TaggedPitchType'])['RV'].sum()

pitch_count = data.groupby(['Pitcher', 'PitcherTeam', 'TaggedPitchType']).size()

result_df = pd.DataFrame({'PV_sum': rv_sum, 'PitchCount': pitch_count})

result_df
```

```python
result_df['PV/100']=(rv_sum/pitch_count)*100
```

```python
result_df.reset_index(inplace=True)
```
